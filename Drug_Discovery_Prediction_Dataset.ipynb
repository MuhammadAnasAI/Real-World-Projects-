# **Drug Discovery Dataset Prediction**

## Dataset Details

### **Dataset Description: Drug Discovery Virtual Screening Dataset**

The **Drug Discovery Virtual Screening Dataset** is a curated and preprocessed collection of molecular, protein, and interaction-specific descriptors designed to facilitate the computational identification of biologically active compounds against specific protein targets. The dataset integrates **chemoinformatics**, **bioinformatics**, and **molecular docking** principles to enable both **ligand-based** and **structure-based virtual screening** workflows.

---

#### **1. Purpose and Scope**

The dataset supports **predictive modeling** for:

* **Classification tasks**: Discriminating between active and inactive compounds against a target protein.
* **Regression tasks**: Estimating binding affinity or docking scores.
* **QSAR/QSPR modeling**: Relating structural features to biological activity.
* **Hit prioritization**: Reducing experimental screening costs by ranking promising candidates.

The dataset captures physicochemical properties, structural descriptors, protein sequence properties, and ligand–protein interaction metrics, making it suitable for **machine learning (ML)**, **deep learning (DL)**, and **statistical modeling** in **drug discovery pipelines**.

---

#### **2. Dataset Composition**

The dataset comprises three feature categories:

##### **A. Compound (Ligand) Features**

Derived from molecular structure analysis using cheminformatics tools such as **RDKit** or **PaDEL-Descriptor**.

| **Feature**                    | **Description**                                    | **Scientific Relevance**                                                                                  |
| ------------------------------ | -------------------------------------------------- | --------------------------------------------------------------------------------------------------------- |
| **compound\_id**               | Unique identifier for each small molecule          | Facilitates traceability and cross-referencing with chemical databases such as PubChem, ChEMBL, or ZINC.  |
| **molecular\_weight (MW)**     | Sum of atomic weights (Da)                         | MW ≤ 500 Da is favorable for oral bioavailability (Lipinski’s Rule). Higher MW may reduce permeability.   |
| **logP**                       | Octanol–water partition coefficient (log scale)    | Measures hydrophobicity; affects membrane permeability and solubility. Optimal range: 1–3 for most drugs. |
| **h\_bond\_donors (HBD)**      | Number of donor groups (N–H, O–H)                  | Crucial for hydrogen bond formation with the protein binding site. ≤5 is preferred for drug-likeness.     |
| **h\_bond\_acceptors (HBA)**   | Number of acceptor atoms (N, O, F with lone pairs) | Facilitates ligand–protein recognition. ≤10 is ideal for oral bioavailability.                            |
| **rotatable\_bonds**           | Number of freely rotating single bonds             | Flexibility affects conformational adaptation to binding sites. Veber’s rule suggests ≤10.                |
| **polar\_surface\_area (PSA)** | Total polar atom surface area (Å²)                 | Influences absorption and blood–brain barrier penetration; ≤140 Å² for oral drugs, ≤60 Å² for CNS drugs.  |
| **compound\_clogP**            | Computationally predicted logP                     | Useful when experimental logP is unavailable; calculated via QSAR/QSPR algorithms.                        |

---

#### **B. Protein Target Features**

Extracted from protein sequence and structure information using tools such as **Biopython**, **Pfeature**, and **PDB-derived data**.

| **Feature**                         | **Description**                                  | **Scientific Relevance**                                                       |
| ----------------------------------- | ------------------------------------------------ | ------------------------------------------------------------------------------ |
| **protein\_id**                     | Unique identifier for each target protein        | Cross-references with UniProt or PDB entries for sequence and structural data. |
| **protein\_length**                 | Number of amino acids in sequence                | Longer proteins may have multiple domains or binding sites.                    |
| **protein\_pI (Isoelectric Point)** | pH at which the protein’s net charge is zero     | Influences protein solubility and ligand binding under physiological pH.       |
| **hydrophobicity**                  | Relative non-polarity of protein surface         | Hydrophobic patches often correspond to ligand-binding pockets.                |
| **binding\_site\_size**             | Size or residue count of the active/binding site | Determines ligand accommodation and binding specificity.                       |

---

#### **C. Interaction and Activity Features**

Obtained from molecular docking simulations, binding affinity measurements, and protein–ligand interaction profiling.

| **Feature**               | **Description**                                                                | **Scientific Relevance**                                                                   |
| ------------------------- | ------------------------------------------------------------------------------ | ------------------------------------------------------------------------------------------ |
| **mw\_ratio**             | Ratio of ligand molecular weight to protein (or binding site) molecular weight | High ratios can indicate steric hindrance; low ratios may reduce interaction stability.    |
| **logp\_pi\_interaction** | Interaction score reflecting hydrophobicity and π–π stacking potential         | Aromatic interactions play a critical role in stabilizing many drug–protein complexes.     |
| **binding\_affinity**     | Quantitative binding strength (Kd, Ki, IC50, or docking score)                 | Lower values indicate stronger interactions; often the main regression target in modeling. |
| **active**                | Binary classification label (1 = active, 0 = inactive)                         | Used as the dependent variable in classification models.                                   |

---

#### **3. Data Sources**

The dataset may integrate data from:

* **Ligand Databases**: PubChem, ChEMBL, ZINC15.
* **Protein Databases**: UniProt, Protein Data Bank (PDB).
* **Docking Simulations**: AutoDock Vina, GOLD, Schrödinger Glide.
* **Experimental Data**: IC50, Ki, and Kd values from biochemical assays.

---

#### **4. Data Generation Workflow**

1. **Compound Selection**: Retrieval of candidate molecules from chemical databases.
2. **Protein Target Preparation**: Sequence acquisition, 3D structure optimization, removal of water molecules, addition of hydrogen atoms.
3. **Feature Extraction**: Generation of molecular descriptors and protein physicochemical properties.
4. **Molecular Docking (Optional)**: Simulation of ligand–protein binding to estimate interaction energies.
5. **Data Cleaning**: Removal of duplicates, outlier detection, handling of missing values.
6. **Normalization**: Feature scaling for machine learning readiness.

---

#### **5. Theoretical Relevance**

* **Lipinski’s Rule of Five** and **Veber’s Rule** are applied to assess drug-likeness.
* Protein–ligand binding is governed by **hydrogen bonding**, **hydrophobic effects**, **electrostatics**, and **van der Waals forces**.
* Combining **ligand descriptors** and **protein features** enhances the ability to generalize predictions across multiple targets (chemogenomics approach).

---

#### **6. Applications**

* **Virtual Screening**: Rapid identification of lead compounds.
* **Drug Repurposing**: Screening existing drugs for new targets.
* **ADMET Prediction**: Integration with absorption, distribution, metabolism, excretion, and toxicity models.
* **QSAR Studies**: Correlation between chemical structure and biological activity.




## Machine Learning Classification Algorithms Details 

1. **Gradient Boosting Classifier**


   Gradient Boosting is an ensemble machine learning algorithm that builds a series of weak learners (usually decision trees) sequentially, where each new tree attempts to correct the errors of the previous ones. It is widely used for classification and regression tasks due to its high predictive accuracy.

---------------

2. **Naive Bayes Classifier**  
   This probabilistic classifier is based on Bayes’ theorem, assuming strong (naive) independence among features. It calculates the posterior probability of classes given the input values and assigns the class with the highest probability. It works well with high-dimensional data and categorical data, despite its simple assumption, and is computationally efficient.

---------------   

3. **Logistic Regression Classifier**  
   A statistical model specifically designed for binary classification, logistic regression models the probability that an input belongs to a particular class using the logistic (sigmoid) function, which outputs values between 0 and 1. It estimates weights for input features to maximize the likelihood of observed class labels. It can also be extended to multinomial classification.

----------------   

4. **XGBoost Classifier**  
   XGBoost (Extreme Gradient Boosting) is an ensemble method based on gradient boosting of decision trees. It builds trees sequentially, each correcting errors of the previous ones, optimizing an objective function with regularization to prevent overfitting. It is highly efficient, supports sparse data, and usually achieves state-of-the-art results on structured data.

-------------------   

5. **Decision Tree Classifier**  
   This classifier uses a tree structure where internal nodes represent decisions on feature values, branches represent outcomes of decisions, and leaf nodes represent class labels. It recursively splits data to maximize class separation based on metrics like Gini impurity or information gain. It is interpretable but prone to overfitting without pruning.

----------------------   

6. **Random Forest Classifier**  
   An ensemble of decision trees, random forest creates multiple trees using random subsets of data and features (bagging and feature randomness). The final class is predicted by majority voting across trees. This approach reduces overfitting and improves generalization compared to a single decision tree.

---------------------------------   

7. **Support Vector Machine (SVM) Classifier**  
   SVM finds the optimal hyperplane that maximally separates classes in the feature space by maximizing the margin between the closest points of different classes (support vectors). It can handle linear and nonlinear classification by using kernel functions to map inputs to higher-dimensional spaces where classes are separable. SVM is effective in high-dimensional spaces and when the number of samples is less than features.

----------------------------
